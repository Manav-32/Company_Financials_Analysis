```python
df.columns=df.columns.str.strip().str.replace(" ","_").str.lower()
df.columns
```

```python
numeric_columns = [ "units_sold", "manufacturing_price", "sale_price", "gross_sales", "sales", "cogs", "profit" ]

for col in numeric_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r"[\$,()]", "", regex=True)
        .str.replace(r"\$-", "0", regex=True)
        .str.replace(r"\-", "0", regex=True)
        .str.strip()
        .astype(float)
    )

```

```python
df["discounts"] = df["discounts"].replace(r"\$-", "0", regex=True).replace(r"[\$,]", "", regex=True)
```

```python
df.head()
```

```python
df["discount_band"].isnull().sum()
```

```python
df["discount_band"].unique()
```

```python
df["discount_band"] = df["discount_band"].str.strip().str.lower()

df["discount_band"].unique()
```

```python
discount_map = {
    'none': 0,
    'low': 1,
    'medium': 2,
    'high': 3
}

df['discount_band'] = df['discount_band'].map(discount_map).fillna(-1).astype(int)
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
monthly_sales = df.groupby("date")["sales"].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind="line")
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
product_sales = df.groupby("product")["sales"].sum()
product_sales.plot(kind="bar",color="lightgreen")
plt.title("Total Sales by Product")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
product_profit = df.groupby("product")["profit"].sum()
product_profit.plot(kind="bar",color="green")
plt.title("Total Profit by Product")
plt.xlabel("Products")
plt.ylabel("Profit")
plt.show()
```

```python
sales_segment = df.groupby("segment")['sales'].sum()

sales_segment.plot(kind="bar",color="blue")
plt.title("Sales by Segment")
plt.xlabel("Segments")
plt.ylabel("Sales")
plt.show()
```

```python
profit_segment = df.groupby("segment")['profit'].sum()

profit_segment.plot(kind="bar",color="red")
plt.title("Profit by Segment")
plt.xlabel("Segments")
plt.ylabel("Profits")
plt.show()
```

```python
country_sales = df.groupby("country")["sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
country_sales.plot(kind="bar", color="orange")
plt.title("Countries by Sales")
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
country_profit = df.groupby("country")["profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
country_profit.plot(kind="bar", color="purple")
plt.title("Countries by Profit")
plt.xlabel("Country")
plt.ylabel("Total Profit")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
df["profit_margin"] = df["profit"] / df["sales"]

plt.figure(figsize=(8,5))
df.boxplot(column="profit_margin", by="segment")
plt.title("Profit Margin Distribution by Segment")
plt.suptitle("")
plt.ylabel("Profit Margin")
plt.show()
```

```python
corr_matrix = df[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
plt.title("Correlation Matrix of Numeric Variables")
plt.show()
```
